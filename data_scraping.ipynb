```python
from bing_image_downloader import downloader

downloader.download("uzbek pilaf plov food",
    limit=100,
    output_dir="/content/pilaf_images",
    adult_filter_off=True,
    force_replace=False,
    timeout=60
)
```

```python
import os

# Papka yo'li (kodda ko'rsatganingizdek)
folder_path = "/content/pilaf_images/uzbek pilaf plov food"

# Fayllarni nomlari bo'yicha saralab olamiz
files = sorted(os.listdir(folder_path))

# Har bir faylni ketma-ketlikda qayta nomlaymiz
for i, filename in enumerate(files):
    # Fayl kengaytmasini aniqlash (.jpg, .png va h.k.)
    file_ext = os.path.splitext(filename)[1]

    # Yangi nom (masalan: 001.jpg, 002.jpg...)
    new_name = f"{i+1:03d}{file_ext}"

    # Eskisini yangisiga o'zgartirish
    os.rename(os.path.join(folder_path, filename), os.path.join(folder_path, new_name))

print("All pictures are now in order!")
```

```python
import os, shutil # Fayllar bilan ishlash uchun kerakli vositalarni yuklaydi.

source_dir = '/content/pilaf_images/uzbek pilaf plov food' # Rasmlar turgan papka manzilini belgilaydi.
target_dir = '/content/drive/MyDrive/food101_dataset/pilaf' # Rasmlar ko'chiriladigan joy manzilini belgilaydi.

os.makedirs(target_dir, exist_ok=True) # Maqsadli papkani yaratadi (agar u bo'lmasa).

for filename in os.listdir(source_dir): # Papkadagi barcha fayllarni birma-bir aylantirib chiqadi.
    src_file = os.path.join(source_dir, filename) # Faylning hozirgi to‘liq manzilini shakllantiradi.
    dst_file = os.path.join(target_dir, filename) # Faylning yangi ko‘chib o‘tadigan to‘liq manzilini shakllantiradi.
    shutil.move(src_file, dst_file) # Faylni eski joyidan yangi joyiga jismonan ko'chiradi.
```
