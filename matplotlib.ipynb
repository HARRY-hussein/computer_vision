```python
import numpy as np
import matplotlib.pyplot as plt
```

# Basic Line Plot

```python
x = np.array([1, 2, 3, 4])         # X o'qi uchun qiymatlarni massiv ko'rinishida yaratish
y = np.array([10, 20, 25, 30])     # Y o'qi uchun mos qiymatlarni massiv ko'rinishida yaratish

plt.plot(x, y)                     # X va Y nuqtalari bo'yicha chiziqli grafikni chizish
plt.title("Basic Line Plot")       # Grafikga "Basic Line Plot" nomli sarlavha qo'yish
plt.xlabel("X Values")             # X o'qiga "X Values" deb nom berish
plt.ylabel("Y Values")             # Y o'qiga "Y Values" deb nom berish
plt.show()                         # Tayyor bo'lgan grafikni ekranda namoyish etish
```

## Line styles and Colors

```python
# Ma'lumotlarni belgilash
x = np.array([1, 2, 3, 4])         # X o'qi qiymatlari (bu misolda plot ichida ishlatilmadi)
y = np.array([3, 8, 1, 10])        # Y o'qi qiymatlari

# Grafikni chizish va uslub berish
plt.plot(y, linestyle='dotted', color='green', marker='o') # Y qiymatlari asosida: chiziq nuqtali, yashil rangda va markerlari aylana shaklida
plt.title("Styled Line Example")   # Grafik sarlavhasi
plt.show()                         # Grafikni ekranga chiqarish
```

# > Multiple Lines in One Graph with NumPy and Malplotlib

```python
# Define data arrays
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([1, 3, 5, 7])

# Plotting the first line with a red color and square markers
plt.plot(x, y1, label='Line 1', color='red', marker='s')

# Plotting the second line with a blue color, dashed linestyle, and circle markers
plt.plot(x, y2, label='Line 2', color='blue', linestyle='--', marker='o')

# Adding chart title and axis labels
plt.title('Multiple Lines Example')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Adding the legend to identify the lines based on their 'label'
plt.legend()

# Display the final plot
plt.show()
```

```python
# 1. Data preparation
x = np.linspace(0, 10, 20)           # Generate 20 points between 0 and 10
y1 = np.sin(x)                       # First line: Sine wave
y2 = np.cos(x)                       # Second line: Cosine wave

# 2. figure size and professional style
plt.figure(figsize=(10, 6), dpi=100)
plt.style.use('seaborn-v0_8-whitegrid') # Gives a clean, professional grid background

# 3. Plot the lines with custom styling
# Using hex colors, specific markers, and distinct line styles
plt.plot(x, y1, label='zync production', color='#1f77b4', marker='o', linewidth=2, linestyle='-')
plt.plot(x, y2, label='zync consumption', color='#ff7f0e', marker='s', linewidth=2, linestyle='--')

# 4. Labels and titles
plt.title("Line Chart Analysis", fontsize=16, fontweight='bold', pad=15)
plt.xlabel("10-day period", fontsize=12)
plt.ylabel("litres", fontsize=12)

# 5. Enhance legibility
plt.legend(fontsize=11, loc='upper right') # Adds a legend to distinguish lines
plt.grid(True, linestyle=':', alpha=0.6)   # Refines the background grid

# 6. Final layout adjustments
plt.tight_layout()
plt.show()
```

> **`BAR CHART with NUMPY and MATPLOTLIB`**



```python
# Define the categories for the X axis and their corresponding values for the Y axis
x = np.array(['Salary', 'Expense', 'Charity', 'Food'])
y = np.array([5, 9, 1, 7])

# Create the bar chart with orange-colored bars
plt.bar(x, y, color='orange')

# Add the title and axis labels
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')

# Display the final plot
plt.show()
```

```python
# 1. Data preparation
categories = ['Jan', 'Feb', 'Mar', 'Apr']
product_a = [20, 35, 30, 45]
product_b = [25, 30, 40, 35]

# Set the positions and width for the bars
x = np.arange(len(categories))  # Label locations
width = 0.35                    # Width of the bars

# 2. Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Create grouped bars
rects1 = ax.bar(x - width/2, product_a, width, label='Product A', color='#4c72b0')
rects2 = ax.bar(x + width/2, product_b, width, label='Product B', color='#dd8452')

# 3. Customizing the chart for a professional look
ax.set_ylabel('Sales (Units)', fontsize=12)
ax.set_title('Monthly Sales Comparison: Product A vs Product B', fontsize=16, pad=20)
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend(fontsize=11)

# Adding value labels on top of the bars
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=10)

autolabel(rects1)
autolabel(rects2)

# Styling
ax.spines['top'].set_visible(False)    # Hide top border
ax.spines['right'].set_visible(False)  # Hide right border
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

# Scatter Plot with NumPy and Malplotlib

```python
# Define the data for the X and Y axes
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 4, 6, 3, 7])

# Create a scatter plot
# color='purple' sets the color of the dots
# s=100 sets the size of the dots
plt.scatter(x, y, color='purple', s=100)

# Add title and axis labels
plt.title('Scatter Plot Example')
plt.xlabel('X')
plt.ylabel('Y')

# Display the plot
plt.show()
```

# Histogram with NumPy and Malplotlib

```python
# Define the data array
data = np.array([3, 4, 5, 6, 8, 8, 8, 9, 10])

# Create a histogram
# bins=5: divides the data into 5 intervals
# color='skyblue': sets the bar color
# edgecolor='black': adds a black border to each bar for better visibility
plt.hist(data, bins=5, color='skyblue', edgecolor='black')

# Add title and axis labels
plt.title("Histogram Example")
plt.xlabel('Values')
plt.ylabel('Count')

# Display the plot
plt.show()
```

# Subplots (Multiple Graphs Together)

```python
# Ma'lumotlarni yaratish
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([3, 6, 9, 12])

# Oyna o'lchamini belgilash: 8 - eni (width), 4 - bo'yi (height)
plt.figure(figsize=(8, 4))

# 1-grafik uchun joy ajratish (1 qator, 2 ustun, 1-o'rin)
plt.subplot(1, 2, 1)
plt.plot(x, y1, color='red')
plt.title('Plot 1')

# 2-grafik uchun joy ajratish (1 qator, 2 ustun, 2-o'rin)
plt.subplot(1, 2, 2)
plt.bar(x, y2, color='blue')
plt.title('Plot 2')

# Grafiklar bir-biriga tegib ketmasligi uchun avtomatik to'g'rilash
plt.tight_layout()
plt.show()
```

```python
# Ma'lumotlarni aniqlash
x = np.array([1, 2, 3, 4])
y = np.array([4, 8, 6, 10])

# Grafikni chizish
plt.plot(x, y, color='green')
plt.title('Save Figure Example')

# Grafikni 'my_plot.png' nomi bilan joriy papkaga saqlash
plt.savefig('my_plot.png')

# Grafikni ekranda ko'rsatish
plt.show()
```

# 3D grafik (Wireframe plot)

```python
import matplotlib.pyplot as plt

# 3D grafiklar uchun zarur bo'lgan kutubxona
from mpl_toolkits.mplot3d import axes3d

# Grafik uslubini belgilash
plt.style.use('_mpl-gallery')

# Ma'lumotlarni yaratish (Matplotlib'ning ichki test ma'lumotlari)
X, Y, Z = axes3d.get_test_data(0.05)

# 3D grafikni boshlash: 'projection': '3d' - bu grafikning 3 o'lchamli bo'lishini bildiradi
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# 3D simli (wireframe) grafikni chizish
# rstride va cstride - qatorlar va ustunlar oralig'ini (qadamini) belgilaydi
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

# O'qlar bo'ylab raqamli belgilarni (tick labels) olib tashlash
ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

# Grafikni ekranda namoyish etish
plt.show()
```

# 3D Scatter Plot

```python
# Grafik uslubini belgilash
plt.style.use('_mpl-gallery')

# Ma'lumotlarni yaratish
np.random.seed(19680801) # Natijalar har safar bir xil bo'lishi uchun (takrorlanuvchanlik)
n = 100                  # Nuqtalar soni
rng = np.random.default_rng() # Tasodifiy sonlar generatori

# X, Y va Z o'qlari uchun tasodifiy qiymatlar
xs = rng.uniform(23, 32, n)    # 23 dan 32 gacha n ta son
ys = rng.uniform(0, 100, n)    # 0 dan 100 gacha n ta son
zs = rng.uniform(-50, -25, n)  # -50 dan -25 gacha n ta son

# Grafikni boshlash: 'projection': '3d' - 3 o'lchamli muhitni ochadi
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# 3D nuqtali grafikni chizish
ax.scatter(xs, ys, zs)

# O'qlar bo'ylab raqamli belgilarni (tick labels) olib tashlash
ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

# Grafikni ekranda ko'rsatish
plt.show()
```
